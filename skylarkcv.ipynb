Initialize the modules and Load the image path list

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import glob
import itertools
import csv

trainlist = glob.glob('train/*.jpg')
validlist= glob.glob('validation/*.JPG')
gcps=[]    
```

Perform the Template Matching to find GCP

```python
#Iterate the trainlist and validlist to perform template matching

for u,v in itertools.product(validlist,trainlist,repeat=1):
    
    #Read the images in colored form
    img =cv2.imread(u,1)
    template=cv2.imread(v,1)
    
    #Use the method TM_CCOEFF_NORMED for matching
    method = eval('cv2.TM_CCOEFF_NORMED')

    # Apply template Matching
    res = cv2.matchTemplate(img,template,method)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)

    center=max_loc
    
    #Compare max_val to a threshold to reject the false identifications
    if(max_val>0.755):
        print(max_val)
        cv2.circle(img,center,150,(255,0,0),3)
        plt.rcParams["figure.figsize"] = [20,10]
        plt.subplot(121),plt.imshow(img)
        plt.title('Possible GCP in image:- '+ u + '  w/  Template type:-'+v), plt.xticks([]), plt.yticks([])
        plt.show()
        gcps.append(str(u)+' GCP- '+str(center))
    else:
        print ("Possibility for a GCP is very low in the image:- "+u+" using the template:- "+v )
```

Export the GCP locations onto a CSV file

```python
csvfile = "gcploc.csv"

with open(csvfile, "w") as output:
    writer = csv.writer(output, lineterminator='\n')
    for val in gcps:
        writer.writerow([val])    
```
